```python
import numpy as np
import astropy.units as u
import matplotlib.pyplot as plt

from MassProfile import MassProfile
```

```python
# Define array of radii
r = np.arange(0.1, 30.1, 0.1)
```

## Milky Way Mass Profiles

```python
# Initialize mass profile object
MW = MassProfile('MW', 0)
```

```python
# Compute enclosed mass for each particle type and total enclosed mass
m_enc = []
for i in range(1, 4):
    m_enc.append(MW.mass_enclosed(i, r))

m_enc.append(MW.mass_enclosed_total(r))
```

```python
# Fit Hernquist mass profile
a = 61.5
m_enc.append(MW.hernquist_mass(r, a))
```

```python
# Plot all profiles
plt.rcParams.update({"font.family": "serif", "figure.dpi": 300})

linestyles = ['-', '-', '-', '-', '--']
labels = ['Halo', 'Disk', 'Bulge', 'Total', f'Hernquist a = {a*u.kpc}']

for i in range(5):
    plt.semilogy(r, m_enc[i], linestyle=linestyles[i], label=labels[i])

plt.title('Mass Profiles for Milky Way')
plt.ylabel(r'Enclosed Mass ($M_\odot$)')
plt.xlabel('Radius (kpc)')
plt.legend(loc='lower right')

plt.savefig('mass_profiles_MW.jpg')
```

## MW Rotation Curve

```python
# Compute rotation curve for each particle type and total rotation curve
vcircs = []
for i in range(1, 4):
    vcircs.append(MW.circular_velocity(i, r))

vcircs.append(MW.circular_velocity_total(r))
```

```python
# Fit Hernquist rotation curve
a = 61.5
vcircs.append(MW.hernquist_vcirc(r, a))
```

```python
# Plot all curves
plt.rcParams.update({"font.family": "serif", "figure.dpi": 300})

linestyles = ['-', '-', '-', '-', '--']
labels = ['Halo', 'Disk', 'Bulge', 'Total', f'Hernquist a = {a*u.kpc}']

for i in range(5):
    plt.semilogy(r, vcircs[i], linestyle=linestyles[i], label=labels[i])

plt.title('Rotation Curve for Milky Way')
plt.ylabel(r'Circular Velocity (kpc $s^{-1}$)')
plt.xlabel('Radius (kpc)')
plt.legend(loc='lower right')

plt.savefig('rotation_curve_MW.jpg')
```
